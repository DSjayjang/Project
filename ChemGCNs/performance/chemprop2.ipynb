# Freesolv

```python
# 배치 32 / 에포크 기본
backbone = 'GCN'
dataset = 'Freesolv'
metric = 'MSE'

seed100_gcn = [ 1.3474748134613037, 1.4423280954360962 ,2.0465750694274902 ,1.1809426546096802 , 1.3439927101135254 ]
seed200_gcn = [ 1.3495995998382568, 1.1330033540725708 ,1.8607776165008545 , 1.8607776165008545 ,  1.119796633720398]
seed300_gcn = [ 1.2251853942871094,  1.0576727390289307, 1.2111797332763672, 0.9403640627861023, 1.0008054971694946]

print(f'backbone: {backbone}')
print(f'metric: {metric}')
print(f'dataset: {dataset}')

# GCN
all_seed_gcn = seed100_gcn + seed200_gcn + seed300_gcn
n = len(all_seed_gcn)
print(n)

all_seed_gcn_avg = np.mean(all_seed_gcn)
std_dev = np.std(all_seed_gcn, ddof = 1)
standard_error = std_dev / math.sqrt(n)
print(f'all_seed_GCN_{metric}:', f'{all_seed_gcn_avg:.3f}')
print(f"all_seed_GCN_se: {standard_error:.3f}\n")
```
