# Freesolv

```python
backbone = 'GCN'
dataset = 'Freesolv'
metric = 'MSE'

seed100_gcn = [1.9830382339597687, 1.2960261547659326, 1.1849035127895085, 1.017817369596226, 1.27285836489146]
seed200_gcn = [1.3095430989903727, 1.0743483896330586, 1.0024774506336123, 1.124471048670491, 1.5403233593656818]
seed300_gcn = [1.9935571865772639, 1.5885684396338275, 1.7987077367587352, 1.2465235192006028, 1.9183831833701097]

print(f'backbone: {backbone}')
print(f'metric: {metric}')
print(f'dataset: {dataset}')

# GAT
all_seed_gcn = seed100_gcn + seed200_gcn + seed300_gcn
n = len(all_seed_gcn)
print(n)

all_seed_gcn_avg = np.mean(all_seed_gcn)
std_dev = np.std(all_seed_gcn, ddof = 1)
standard_error = std_dev / math.sqrt(n)
print(f'all_seed_GCN_{metric}:', f'{all_seed_gcn_avg:.3f}')
print(f"all_seed_GCN_se: {standard_error:.3f}\n")
```
