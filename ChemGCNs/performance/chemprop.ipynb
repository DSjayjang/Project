# Freesolv

```python
# 확인 완료
backbone = 'GCN'
dataset = 'Freesolv'
metric = 'MSE'

seed100_gcn = [0.5598945021629333, 0.8200034499168396, 0.6253548860549927, 1.6444346904754639, 2.5263595581054688]
seed200_gcn = [0.8748635649681091, 1.1275814771652222, 1.1258677244186401, 1.2503942251205444, 1.0977809429168701]
seed300_gcn = [0.7454223036766052, 0.6878228187561035, 0.688782274723053, 0.5315753221511841, 1.1988627910614014]

print(f'backbone: {backbone}')
print(f'metric: {metric}')
print(f'dataset: {dataset}')

# GCN
all_seed_gcn = seed100_gcn + seed200_gcn + seed300_gcn
n = len(all_seed_gcn)
print(n)

all_seed_gcn_avg = np.mean(all_seed_gcn)
std_dev = np.std(all_seed_gcn, ddof = 1)
standard_error = std_dev / math.sqrt(n)
print(f'all_seed_GCN_{metric}:', f'{all_seed_gcn_avg:.3f}')
print(f"all_seed_GCN_se: {standard_error:.3f}\n")
```
