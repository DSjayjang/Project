```python
import os
import pandas as pd

import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
dataset = 'solubility' # freesolv / esol / lipo / scgas / solubility
save_dir = r'..\results_figure\scatter'
```

# MSE & MAE for Train dataset

```python
# MSE
df_train_MSE = pd.read_csv(r'..\results_datasets\result_train_' + dataset + '_mse'  + '.csv')
print(df_train_MSE.shape[1])

# # MAE
# df_train_MAE = pd.read_csv(r'..\results_datasets\result_train_' + dataset + '_mae' + '.csv')
# print(df_train_MAE.shape[1])
```

```python
# MSE 컬럼명 추가
columns = ['target', 'fitted'] + [f'feature{i}' for i in range(1, df_train_MSE.shape[1]-1)]
df_train_MSE.columns = columns

# # MAE 컬럼명 추가
# columns = ['target', 'fitted'] + [f'feature{i}' for i in range(1, df_train_MAE.shape[1]-1)]
# df_train_MAE.columns = columns
```

```python
# R^2
r2_mse = r2_score(df_train_MSE['target'], df_train_MSE['fitted'])
print('r2_mse:', r2_mse)

# adjusted R^2
adjusted_r2_mse = ((len(df_train_MSE) - 1) * r2_mse - df_train_MSE.shape[1]) / (len(df_train_MSE) - df_train_MSE.shape[1] - 1)
print('adjusted r2_mse:', adjusted_r2_mse)
```

```python
if dataset in ['freesolv', 'esol', 'lipo', 'logvp', 'scgas', 'solubility']:
    s = 15
elif dataset in ['vd', 'logp', 'logvp_old']:
    s = 25
```

# MSE

```python
from matplotlib.colors import ListedColormap
pastel_cmap = ListedColormap(["black"])

plt.figure(figsize = (8, 6))

scatter = plt.scatter(df_train_MSE['fitted'], df_train_MSE['target'],
                      c=df_train_MSE['target'], s = s,
                      cmap = pastel_cmap)

# 대각선 추가
min_val = min(df_train_MSE['target'].min(), df_train_MSE['fitted'].min()) - 20
max_val = max(df_train_MSE['target'].max(), df_train_MSE['fitted'].max()) + 20
plt.plot([min_val, max_val], [min_val, max_val], color='red')

if dataset == 'freesolv':
    plt.xlim(-15, 6)
    plt.ylim(-15, 6)
    plt.xticks(range(-15, 9, 3))
    plt.yticks(range(-15, 9, 3))
elif dataset == 'esol':
    plt.xlim(-12, 4)
    plt.ylim(-12, 4)
    plt.xticks(range(-12, 5, 2))
    plt.yticks(range(-12, 5, 2))
elif dataset == 'lipo':
    plt.xlim(-4, 8)
    plt.ylim(-4, 8)
    plt.xticks(range(-4, 10, 2))
    plt.yticks(range(-4, 10, 2))
elif dataset == 'scgas':
    plt.xlim(-12, 7)
    plt.ylim(-12, 7)
    plt.xticks(range(-12, 8, 2))
    plt.yticks(range(-12, 8, 2))
elif dataset == 'solubility':
    plt.xlim(-16, 6)
    plt.ylim(-16, 6)
    plt.xticks(range(-16, 7, 2))
    plt.yticks(range(-16, 7, 2))

plt.xlabel("Fitted Values", fontweight = 'bold')
plt.ylabel("Observations", fontweight = 'bold')

plt.text(0.05, 0.90, rf'$R{{^2}}$   = {r2_mse:.4f}',
         transform = plt.gca().transAxes,
         fontsize = 10,)
# plt.text(0.05, 0.86, rf'$R_{{adj}}^2$ = {adjusted_r2_mse:.4f}',
#          transform = plt.gca().transAxes,
#          fontsize = 10,)

plt.grid(alpha = 0.2)

# # 저장
# file_name = f'scatter_fitted_mse_{dataset}.png'
# file_path = os.path.join(save_dir, file_name)
# print(file_path)
# plt.savefig(file_path, dpi = 300, bbox_inches = 'tight')

plt.show()
```
