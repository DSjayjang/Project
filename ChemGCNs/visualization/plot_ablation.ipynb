# Results

```python
datasets = ['Freesolv', 'ESOL', 'Lipo']
models = ['GCN', 'GAT', 'EGCN', 'Ours']
colors = ['#cacaca', '#C8A26B', '#94602e', '#b23e49']

# MSE #
performance_mse = np.array([[8.751, 6.480, 5.636, 0.996],
                            [3.349, 3.104, 0.892, 0.411],
                            [1.206, 1.091, 1.126, 0.580]])
performance_r2 = np.array([[0.350, 0.439, 0.541, 0.869],
                           [0.214, 0.212, 0.792, 0.869],
                           [0.207, 0.293, 0.284, 0.622]])
```

```python
x = np.arange(len(datasets))
width = 0.2

fig, ax = plt.subplots(figsize = (8, 6))

for i in range(len(models)):
    ax.bar(x + i * width - width * 1.5,
           performance_mse[:, i],
           width, label = models[i], alpha = 0.9, color = colors[i])

ax.set_xticks(x)
ax.set_xticklabels(datasets)
ax.set_title('MSE')
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()

# # 저장
# file_name = f'bar_mse.png'
# file_path = os.path.join(save_dir, file_name)
# plt.savefig(file_path, dpi = 300, bbox_inches = 'tight')

plt.show()
```

```python
x = np.arange(len(datasets))
width = 0.2

fig, ax = plt.subplots(figsize = (8, 6))

for i in range(len(models)):
    ax.bar(x + i * width - width * 1.5,
           performance_r2[:, i],
           width, label = models[i], alpha = 0.85, color = colors[i])

ax.set_xticks(x)
ax.set_xticklabels(datasets)
ax.set_title('$R_{adj}^{2}$')
ax.legend(loc = 'upper left')
ax.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()

# # 저장
# file_name = f'bar_r2.png'
# file_path = os.path.join(save_dir, file_name)
# plt.savefig(file_path, dpi = 300, bbox_inches = 'tight')

plt.show()
```
