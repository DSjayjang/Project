```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset_freesolv = 'freesolv' # freesolv / esol / lipo / logvp / VP
df_freesolv = pd.read_csv(f'..\datasets\{dataset_freesolv}.csv')

dataset_esol = 'esol' # freesolv / esol / lipo / logvp / VP
df_esol = pd.read_csv(f'..\datasets\{dataset_esol}.csv')

save_dir = r'..\results_figure\histogram'
```

```python
fig = plt.figure(figsize=(6.69, 3.0))
ax1 = fig.add_subplot(121)

sns.histplot(df_freesolv['freesolv'], color = 'darkgray', edgecolor='white', bins=20,  alpha = 1, stat='density')
sns.kdeplot(df_freesolv['freesolv'], ax=ax1, linewidth=1.0, color='slategray')

ax1.set_xlabel('')
ax1.set_ylabel('')
ax1.set_xticks(range(df_freesolv['freesolv'].min().astype(int), df_freesolv['freesolv'].max().astype(int)+5, 10))
ax1.set_yticks(np.arange(0, 0.15, 0.03))
ax1.tick_params(axis='both', labelsize=8)

ax2 = fig.add_subplot(122)

sns.histplot(df_esol['logp'], color = 'darkgray', edgecolor='white', bins=20,  alpha = 1, stat='density')
sns.kdeplot(df_esol['logp'], ax=ax2, linewidth=1.0, color='slategray')

ax2.set_xlabel('')
ax2.set_ylabel('')
ax2.set_xticks(range(df_esol['logp'].min().astype(int), df_esol['logp'].max().astype(int)+5, 5))
ax2.set_yticks(np.arange(0, 0.22, 0.05))
ax2.tick_params(axis='both', labelsize=8)

fig.text(0.30, -0.05, "(a) Freesolv", ha="center", fontsize=11)
fig.text(0.70, -0.05, "(b) ESOL", ha="center", fontsize=11)

# # 저장
# file_name = '4.pdf'
# file_path = os.path.join(save_dir, file_name)
# print(file_path)
# plt.savefig(file_path, dpi=300, bbox_inches = "tight")

plt.show()
```

# Vapor Pressure & Solubility

```python
dataset_vp = 'scgas'
df_vp = pd.read_csv(f'..\datasets\{dataset_vp}.csv')

dataset_solubility = 'solubility_only3'
df_solubility = pd.read_csv(f'..\datasets\{dataset_solubility}.csv')

save_dir = r'..\results_figure\histogram'
```

```python
fig = plt.figure(figsize=(6.69, 3.0))
ax1 = fig.add_subplot(121)

sns.histplot(df_vp['logvp'], color = 'darkgray', edgecolor='white', bins=20,  alpha = 1, stat='density')
sns.kdeplot(df_vp['logvp'], ax=ax1, linewidth=1.0, color='slategray')

ax1.set_xlabel('')
ax1.set_ylabel('')
ax1.set_xticks(range(df_vp['logvp'].min().astype(int), df_vp['logvp'].max().astype(int)+5, 5))
ax1.set_yticks(np.arange(0, 0.22, 0.04))
ax1.tick_params(axis='both', labelsize=8)

ax2 = fig.add_subplot(122)

sns.histplot(df_solubility['solubility'], color = 'darkgray', edgecolor='white', bins=20,  alpha = 1, stat='density')
sns.kdeplot(df_solubility['solubility'], ax=ax2, linewidth=1.0, color='slategray')

ax2.set_xlabel('')
ax2.set_ylabel('')
ax2.set_xticks(range(df_solubility['solubility'].min().astype(int), df_solubility['solubility'].max().astype(int)+5, 5))
ax2.set_yticks(np.arange(0, 0.22, 0.05))
ax2.tick_params(axis='both', labelsize=8)

fig.text(0.30, -0.05, "(a) Vapor Pressure", ha="center", fontsize=11)
fig.text(0.73, -0.05, "(b) Solubility", ha="center", fontsize=11)

# # 저장
# file_name = '20.pdf'
# file_path = os.path.join(save_dir, file_name)
# print(file_path)
# plt.savefig(file_path, dpi=300, bbox_inches = "tight")

plt.show()
```
