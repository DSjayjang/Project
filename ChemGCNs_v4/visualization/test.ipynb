```python
import os
import pandas as pd

import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
df_train_MSE = pd.read_csv(r'..\results\result_test.csv')
df_train_MSE
```

```python
r2_mse = r2_score(df_train_MSE['target'], df_train_MSE['pred'])
print('r2_mse:', r2_mse)
```

```python
from matplotlib.colors import ListedColormap
pastel_cmap = ListedColormap(["black"])
s = 15

plt.figure(figsize = (8, 6))

scatter = plt.scatter(df_train_MSE['pred'], df_train_MSE['target'],
                      c=df_train_MSE['target'], s = s,
                      cmap = pastel_cmap)

# 대각선 추가
min_val = min(df_train_MSE['target'].min(), df_train_MSE['pred'].min()) - 20
max_val = max(df_train_MSE['target'].max(), df_train_MSE['pred'].max()) + 20
plt.plot([min_val, max_val], [min_val, max_val], color='red')

plt.xlim(-12, 7)
plt.ylim(-12, 7)
plt.xticks(range(-12, 8, 2))
plt.yticks(range(-12, 8, 2))

plt.xlabel("Fitted Values", fontweight = 'bold')
plt.ylabel("Observations", fontweight = 'bold')

plt.text(0.05, 0.90, rf'$R{{^2}}$   = {r2_mse:.4f}',
         transform = plt.gca().transAxes,
         fontsize = 10,)
# plt.text(0.05, 0.86, rf'$R_{{adj}}^2$ = {adjusted_r2_mse:.4f}',
#          transform = plt.gca().transAxes,
#          fontsize = 10,)

plt.grid(alpha = 0.2)

# # 저장
# file_name = f'scatter_fitted_mse_{dataset}.png'
# file_path = os.path.join(save_dir, file_name)
# print(file_path)
# plt.savefig(file_path, dpi = 300, bbox_inches = 'tight')

plt.show()
```
