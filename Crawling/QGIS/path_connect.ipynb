```python
import numpy as np
import pandas as pd
```

```python
Mday = 30
# length_from = 0
# length_to = 1128
```

```python
df = pd.read_csv(fr'..\naver_paths_matrix_{Mday}.csv')
df
```

```python
# 버스 대수만큼 column 복제해야 함
df_map = pd.read_csv(fr'df_map_{Mday}.csv')

expanded_cols = []
df_map['idx'] = df_map['idx'].astype(str)

for col in df.columns:
    repeat_n = df_map.loc[df_map['idx'] == col, 'num_bus'].values[0]

    for i in range(repeat_n):
        expanded_cols.append(df[col].rename(f'{col}_{i+1}'))

df_temp = pd.concat(expanded_cols, axis = 1)
```

```python
# 경로 이어주기
cols = df_temp.columns
df_new = pd.DataFrame()

for col in cols:
    s = df_temp[col].astype('string')
    left = s.iloc[:-1].reset_index(drop=True)
    right = s.iloc[1:].reset_index(drop=True)
    df_new[col] = left + "/" + right

print(df_new)
```

```python
results = pd.DataFrame(columns = ['path'])

for col in cols:
    temp = pd.DataFrame({'path': list(set(df_new[col]))})
    results = pd.concat([results, temp], axis = 0, ignore_index = True)
results
```

```python
df_count = pd.DataFrame(results['path'].value_counts(ascending = False)).reset_index()

df_count['start_path'] = df_count['path'].str.split('/', expand = True)[0]
df_count['end_path'] = df_count['path'].str.split('/', expand = True)[1]
df_count['start_path_lat'] = df_count['start_path'].str.split(',', expand = True)[1]
df_count['start_path_lng'] = df_count['start_path'].str.split(',', expand = True)[0]
df_count['end_path_lat'] = df_count['end_path'].str.split(',', expand = True)[1]
df_count['end_path_lng'] = df_count['end_path'].str.split(',', expand = True)[0]
df_count.shape
```

```python
# 저장
df_count.to_csv(f'df_result_{Mday}.csv', index = True, index_label = 'index')
```
