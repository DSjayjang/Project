```python
import numpy as np
import pandas as pd
```

# 버스 카운트

```python
for day in range(1, 31):
    for hour in range(0, 23):
        try:
            # 날짜별로 저장한 df 다시 불러오기
            df_group = pd.read_csv(fr'..\datasets\geo_group_midtofinal\geo_group_midtofinal_{day}_{hour}.csv', encoding = 'cp949')
            df_group.groupby(list(df_group.columns)).size().reset_index(name = 'count')

            path = df_group.value_counts().reset_index(name = 'count').sort_values(by='count', ascending=False)

            path['num_bus'] = np.ceil(path['count'] / 20).astype(int)

            # idx, bus 매칭시켜서 df로 저장해두기
            # 나중에 컬럼을 bus 대수만큼 복제해야함
            idx_bus = path['num_bus'].to_dict()
            df_map = pd.DataFrame(list(idx_bus.items()), columns = ['idx', 'num_bus'])

            df_map.to_csv(fr'..\datasets\df_map\df_map_{day}_{hour}.csv', index = False)

            df_path = pd.DataFrame()
            df_path['idx'] = path.index
            df_path[['mid_lat', 'mid_lng', 'end_lat', 'end_lng', 'mid_time', 'end_time']] = path[['mid_lat', 'mid_lng', 'end_lat', 'end_lng', '중간집결 기한', '최종도착 기한']]
            df_path.to_csv(fr'..\datasets\df_path\df_path_{day}_{hour}.csv', index = False, encoding = 'cp949')

        except FileNotFoundError:
            print(f"⚠️ 파일 없음: geo_group_midtofinal_{day}_{hour}.csv → 건너뜀")
            continue
```

# 집단 중간 > 최종 필요한 총 버스 대수

```python
all_num_bus = []

for day in range(1, 31):
    day_num_bus = []

    for hour in range(0, 23):
        try:
            # 날짜별로 저장한 df 다시 불러오기
            df_map = pd.read_csv(fr'..\datasets\df_map\df_map_{day}_{hour}.csv', encoding = 'cp949')
            # print(df_map['num_bus'])
            bus = int(np.sum(df_map['num_bus']))
            print(f'M+{day}-{hour}시 bus:', bus)
        except FileNotFoundError:
            print(f"⚠️ 파일 없음: geo_group_midtofinal_{day}_{hour}.csv → 건너뜀")
            continue    
        day_num_bus.append(bus)
    print('day_num_bus:', day_num_bus)

    all_num_bus.append(int(np.sum(day_num_bus)))
    print('all_num_bus:', all_num_bus)

print(f'총 필요버스 수: {np.sum(all_num_bus)}대')
```
