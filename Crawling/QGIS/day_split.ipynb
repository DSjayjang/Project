```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'..\datasets\bottle.csv', encoding = 'cp949')
df
```

```python
df_group = df.loc[df['이동방법'] == '집단'].reset_index(drop = True)

num = df_group['순번']
df_group = df_group.drop(columns = ['출발지 주소', '중간집결지 주소', '최종 도착지 주소', '이동방법', 'start_lat', 'start_lng', '순번'])
df_group
```

```python
df.groupby(list(df_group.columns)).size().reset_index(name = 'count')
```

```python
path = df_group.value_counts().reset_index(name = 'count').sort_values(by='count', ascending=False)
path
```

```python
path['num_bus'] = np.ceil(path['count'] / 20).astype(int)
path
```

```python
df_path = path.loc[path.index.repeat(path['num_bus'].astype(int))].reset_index(drop = False)
df_path.drop(columns = ['count', 'num_bus'], inplace = True)
df_path
```

```python
df_path.to_csv('df_path.csv', index = False, encoding = 'cp949')
```
