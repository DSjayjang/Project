```python
df_group = df.loc[df['이동방법'] == '집단'].reset_index(drop = True)

num = df_group['순번']
df_group = df_group.drop(columns = ['출발지 주소', '중간집결지 주소', '최종 도착지 주소', '이동방법', 'start_lat', 'start_lng', '순번'])
df_group
```

# 날짜 & 시간 나눠서 저장

```python
# 하기 전에,
# 첫 날 도착 시간이 10시인건 제외
df_group = df_group[df_group['최종도착 기한'] != '0+1일 10시']
```

```python
for day in range(1,31):
    for hour in range(0, 23):
        key = f'0+{day}일 {hour}시'
        temp = df_group[df_group['중간집결 기한'].str.contains(key, regex = False)]

        if temp.empty:
            continue

        temp.to_csv(fr'..\datasets\geo_group_midtofinal\geo_group_midtofinal_{day}_{hour}.csv', index = False, encoding = 'cp949')
```
