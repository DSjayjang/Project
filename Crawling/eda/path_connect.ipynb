```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\user\Desktop\연구\5. 국방부 용역과제\naver_paths_matrix_0_10.csv')
df
```

```python
# 경로 이어주기
cols = df.columns
df_new = pd.DataFrame()

for col in cols:
    df_new[col] = df[col].iloc[:-1].reset_index(drop=True) + "/" + df[col].iloc[1:].reset_index(drop=True)

print(df_new)
```

```python
results = pd.DataFrame(columns = ['num'])

for col in cols:
    temp = pd.DataFrame({'num': list(set(df_new[col]))})
    results = pd.concat([results, temp], axis = 0, ignore_index = True)
results
```

```python
df_count = pd.DataFrame(results['num'].value_counts(ascending = False)).reset_index()
df_count['start_path'] = df_count['num'].str.split('/', expand = True)[0]
df_count['end_path'] = df_count['num'].str.split('/', expand = True)[1]
```

```python
df_count['start_path_lat'] = df_count['start_path'].str.split(',', expand = True)[1]
df_count['start_path_lng'] = df_count['start_path'].str.split(',', expand = True)[0]
df_count['end_path_lat'] = df_count['end_path'].str.split(',', expand = True)[1]
df_count['end_path_lng'] = df_count['end_path'].str.split(',', expand = True)[0]
```

```python
df_count
```

```python
df_count['geometry'] = 'LINESTRING (' + df_count['start_path_lng'] + ' ' + df_count['start_path_lat'] + ', ' + df_count['end_path_lng'] + ' ' + df_count['end_path_lat'] + ')'
df_count['geometry'][0]
```

```python
df_result = df_count.drop(columns = ['num', 'start_path', 'end_path', 'start_path_lat', 'start_path_lng', 'end_path_lat', 'end_path_lng'])
df_result
```

```python
# 저장
df_result.to_csv('df_result.csv')
```
