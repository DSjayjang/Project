```python
import time
import numpy as np
import pandas as pd

import requests
```

```python
df_geo_new = pd.read_csv(r'..\datasets\df_geo_new.csv') # 위경도 좌표
df_transformed = pd.read_csv(r'..\datasets\transformed.csv', ) # 위경도 좌표

print(df_geo_new.head())
print(df_transformed.head())
```

```python
print(df_geo_new.shape)
print(df_transformed.shape)
```

```python
df_transformed.dropna()
```

```python
df_merge = pd.merge(df_geo_new, df_transformed, on = 'num', how = 'left')
df_merge
```

```python
for col in ['start_lat', 'start_lng', 'mid_lat', 'mid_lng', 'end_lat', 'end_lng']:
    df_merge[col] = df_merge[f'{col}_x'].fillna(df_merge[f'{col}_y'])
df_merge.drop(columns = ['start_lat_x', 'start_lng_x', 'mid_lat_x', 'mid_lng_x', 'end_lat_x', 'end_lng_x', 'start_lat_y', 'start_lng_y', 'mid_lat_y', 'mid_lng_y', 'end_lat_y', 'end_lng_y'], inplace = True)
df_merge
```

```python
# 저장
save_dir = r'..\datasets\\'
df_merge.to_csv(f'{save_dir}df_geo_new.csv', index = False)
```
